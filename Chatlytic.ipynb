```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Aniket.txt', 'r', encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'#for the date and time part
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
# Strip any trailing spaces from the 'message_date' column
df['message_date'] = df['message_date'].str.strip()

# Parse the dates with the corrected format
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %H:%M -')

# Rename the column
df.rename(columns={'message_date': 'date'}, inplace=True)

# Display the head of the DataFrame
df.head()
# df.shape
```

## Separating user name and message sent by him/her

```python
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)

df.head()
```
